```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
page = urlopen("https://weimergeeks.com/examples/scraping/example1.html")
soup = BeautifulSoup(page, "html.parser")
print(soup.h1)
```

When you extract information from an object with a BeautifulSoup command, sometimes you get a single Tag object, and sometimes you get a Python list (similar to an array in JavaScript) of Tag objects. The way you treat that extracted information will be different depending on whether it is one item or a list (usually, but not always, containing more than one item).

```python
from bs4 import BeautifulSoup
html = "<h1>Hello</h21"
soup = BeautifulSoup(html,"html.parser")
tag = soup.find("h1")# find() returns the First match
print(type(tag))
print(tag.text)
```

```python
html ="<p>look what u made me do</p><p> I want to eat</p><p>Noodles</p>"
soup = BeautifulSoup(html,"html.parser")
tags = soup.find_all("p")
print(type(tags))
print(len(tags))
for tag in tags:
    print(tag.text)
```

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup

# URL
url = "https://www.chess.com"

# Add headers (some sites block requests without them) because it pretends to be like a browser
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
response = urlopen(req)

# Read and decode HTML
#html = response.read().decode("utf-8")
soup = BeautifulSoup(response, "html.parser")
print("Page title:", soup.title.string)
links = soup.find_all("a",  href= True)
for link in links[:10]:
    print(link["href"])
```

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup

url = "https://kathmandupost.com/"
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
response = urlopen(req)

# Parse the HTML
soup = BeautifulSoup(response, "html.parser")

# Print the page title
print("Page title:", soup.title.string if soup.title else "No title found")

# Try to get the first headline safely
headline = soup.find("h2", class_="headline")
if headline:
    print("Headline:", headline.text)
else:
    print("Headline not found")

print("\nFirst 10 links on the page:")
links = soup.find_all("a", href=True)
for link in links[:10]:
    print(link["href"])
```
